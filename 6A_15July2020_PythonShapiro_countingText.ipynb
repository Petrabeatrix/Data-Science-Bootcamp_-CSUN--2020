### Example:
  * Problem: Count the occurences of each lettr in the Gettysburg Address 
       + Subproblem: Count the number of letters in a single stringM

```python
#Version1 
# counting letters in each row 

def countLetters(somestring): 
    line=somestring.lower() # everything is lower case
    counts=26*[0] # create a list of 26 zeros
    # go through each letter in the line
    # figure out if it is an a, b, c, etc.
    # if it is a, increment counts[0]
    # if it is b, increment counts[1]
    # if it is c, increment counts[2]
    # ... etc
    for character in line:
        #  check to see if the character is text
        #  ord(character)-ord('a') tells has how many characters
        #  after "a" each character in the ascii table
        if character.isalpha(): # returns true if a,b,c,...,z
            counts[ord(character)-ord('a')]+=1
    return(counts)
countLetters(g[15])  # counts 14.line of txt
```

* use join, 
+ enables to count whole letters in txt

```python
getty ="".join(g) # join method, new function
print(getty)
```

```python
print(countLetters(getty))
```
