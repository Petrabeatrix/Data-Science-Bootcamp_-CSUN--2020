(e) Perform 100 train/test splits and collect the errors as a function of degree. You
will have 100 errors for each degree.

```python
# perform 100 train/test using for loop, polyfit() and mean_squared_error()

xtrain,xtest,ytrain,ytest = train_test_split(x,y)
fit=np.polyfit(xtrain,ytrain,1) # returns polynomial coefficients
ypredicted=np.polyval(fit, xtest) # pass in coefficents and sequence of x vals, output is sequence of yvals
mean_squared_error(ytest,ypredicted) # MSE, sum squared errors
```

```python
errors=[]
for repeat in range(100):
    xtrain,xtest,ytrain,ytest=train_test_split(x,y)
    fit=np.polyfit(xtrain,ytrain,1) # returns polynomial coefficients
    ypredicted=np.polyval(fit, xtest) # pass in coefficents and sequence of x vals, output is sequence of yvals
    MSE=mean_squared_error(ytest,ypredicted) # MSE, sum squared errors
    errors.append(MSE)
```

```python
#  (f) determine mean and standard deviation of errors
np.mean(errors), np.std(errors)
```

(g) Make a plot (you may want to use a logarithmic scale on the y-axis) of the error as a function of degree. (As an example, see the middle line, with the dots, below.)

```python
plt.scatter(x, y)
for repeat in range(100):
    xtrain,xtest,ytrain,ytest=train_test_split(x, y)   # split
    b,a=np.polyfit(xtrain,ytrain,1)                   # linear fit
    plt.plot([0,16],[a+b*0, a+b*16])
```

```python
def fit_a_polynomial(X,Y,degree):
    xtrain,xtest,ytrain,ytest=train_test_split(X,Y)
    fit=np.polyfit(xtrain,ytrain,degree) # polynomial coefficients of degree degree
    ypredicted=np.polyval(fit,xtest)     # input: coefficients (ouput of polyfit); output: yvalues
    error=mean_squared_error(ytest,ypredicted)
    return(error)
fit_a_polynomial(x,y,1)
```

```python
[fit_a_polynomial(x,y,2) for j in range(10)] # MSE for 10 runs of degree 2
```

```python
degrees=list(range(1,15))
means=[]
sigmas=[]
for deg in degrees:
    errors=[]
    for repeat in range(200):
        errors.append(fit_a_polynomial(x,y,deg))
    means.append(np.mean(errors)) # add the mean error for next degree
    sigmas.append(np.std(errors)) # add the std error for next degree
    
print(means)
print(sigmas)
```
